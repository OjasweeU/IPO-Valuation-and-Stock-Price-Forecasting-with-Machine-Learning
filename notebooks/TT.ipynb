```python
def add_features(df):
    # Ensure that the 'Date' column is in datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Set the 'Date' column as the index
    df.set_index('Date', inplace=True)
    
    # Create the new features
    df['7_day_MA'] = df['Close'].rolling(window=7).mean()
    df['30_day_MA'] = df['Close'].rolling(window=30).mean()
    df['Daily_Return'] = df['Close'].pct_change()
    df['Volatility'] = df['Daily_Return'].rolling(window=7).std()
    df['Monthly_Return'] = df['Close'].resample('M').ffill().pct_change()
    
    # Reset the index if necessary
    df.reset_index(inplace=True)
    
    return df

# Apply feature engineering to each dataset
tata_tech = add_features(tata_tech)
tech_mahindra = add_features(tech_mahindra)
infosys = add_features(infosys)
wipro = add_features(wipro)
hcl_tech = add_features(hcl_tech)

# Display the updated dataframe
print(tata_tech.head())
```

```python
import pandas as pd
import mplfinance as mpf
import matplotlib.pyplot as plt

# Load the CSV files
files = {
    'TATA TECH': 'TataTech One Yr data.csv',
    'TECHM': 'TechM One Yr data.csv',
    'INFY': 'Infosys one YR data.csv',
    'WIPRO': 'Wipro One Yr data.csv',
    'HCLTECH': 'HCLTech One Yr data.csv'
}

# Read the CSV files into DataFrames
data = {name: pd.read_csv(file) for name, file in files.items()}

# Process the data: Convert dates and set index
for name, df in data.items():
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    df.sort_index(inplace=True)
    
    # Calculate moving averages
    df['MA20'] = df['Close'].rolling(window=20).mean()  # 20-day moving average
    df['MA50'] = df['Close'].rolling(window=50).mean()  # 50-day moving average

# Plotting candlestick charts with moving averages for each stock
for name, df in data.items():
    # Add moving averages to the plot
    moving_averages = [mpf.make_addplot(df['MA20'], color='blue', label='MA 20'),
                       mpf.make_addplot(df['MA50'], color='red', label='MA 50')]
    
    mpf.plot(df, type='candle', title=f'{name} Stock Price with Moving Averages',
             style='charles', addplot=moving_averages, volume=True)
    plt.show()
```

```python
# Outlier Detection using IQR
def detect_outliers(df):
    Q1 = df['Close'].quantile(0.25)
    Q3 = df['Close'].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df['Close'] < (Q1 - 1.5 * IQR)) | (df['Close'] > (Q3 + 1.5 * IQR))]
    return outliers

# Detect outliers in Tata Technologies dataset
tata_tech_outliers = detect_outliers(tata_tech)
print(tata_tech_outliers)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Standardization (Min-Max Scaling)
scaler = MinMaxScaler()

# Applying Min-Max Scaling to relevant features
tata_tech[['Close_Scaled', '7_day_MA_Scaled']] = scaler.fit_transform(tata_tech[['Close', '7_day_MA']])
```

```python
# Normalization using z-score for better distribution alignment
tata_tech['Close_Normalized'] = (tata_tech['Close'] - tata_tech['Close'].mean()) / tata_tech['Close'].std()
```

```python
# Data Merge (Combining datasets for a holistic view)
# Example: Merging Tata Tech with Tech Mahindra on Date

# Ensure date format consistency before merging
tata_tech['Date'] = pd.to_datetime(tata_tech['Date'])
tech_mahindra['Date'] = pd.to_datetime(tech_mahindra['Date'])

# Merge the datasets on the Date column
merged_data = pd.merge(tata_tech, tech_mahindra, on='Date', suffixes=('_TATA', '_TECHM'))

# Display the merged data
print(merged_data.head())
```

```python
# Correlation Analysis to understand relationships between features
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix for Tata Technologies
correlation_matrix = tata_tech.corr()
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix for Tata Technologies")
plt.show()
```

```python
# Feature Selection: Based on correlation and importance

# Selecting features with correlation higher than a threshold
threshold = 0.7
strong_corrs = correlation_matrix['Close'].abs().sort_values(ascending=False)
selected_features = strong_corrs[strong_corrs > threshold].index
print(f"Selected Features: {list(selected_features)}")

# These selected features can be used as inputs to the models
X_selected = tata_tech[selected_features].drop(columns=['Close'])
y = tata_tech['Close']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Define a pipeline with imputation and SVR
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('svr', SVR(kernel='rbf'))
])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Fit the pipeline
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

# Model Evaluation
print("SVR RMSE:", mean_squared_error(y_test, y_pred, squared=False))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Define a pipeline with imputation and Random Forest
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('rf', RandomForestRegressor(n_estimators=100, random_state=42))
])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Fit the pipeline
pipeline.fit(X_train, y_train)
y_pred_rf = pipeline.predict(X_test)

# Model Evaluation
print("Random Forest RMSE:", mean_squared_error(y_test, y_pred_rf, squared=False))
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
import numpy as np

# Reshape data for LSTM
X_lstm = tata_tech[['Close_Scaled']].values.reshape(-1, 1)
y_lstm = tata_tech['Close'].values

X_lstm_train, X_lstm_test = X_lstm[:-50], X_lstm[-50:]
y_lstm_train, y_lstm_test = y_lstm[:-50], y_lstm[-50:]

# Reshape input to be [samples, time steps, features]
X_lstm_train = np.reshape(X_lstm_train, (X_lstm_train.shape[0], 1, X_lstm_train.shape[1]))
X_lstm_test = np.reshape(X_lstm_test, (X_lstm_test.shape[0], 1, X_lstm_test.shape[1]))

# LSTM Model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(1, 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_lstm_train, y_lstm_train, epochs=20, batch_size=32)

# Predictions
y_pred_lstm = model.predict(X_lstm_test)
print("LSTM RMSE:", mean_squared_error(y_lstm_test, y_pred_lstm, squared=False))

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# Load the CSV files
files = {
    'TATA TECH': 'TataTech One Yr data.csv',
    'TECHM': 'TechM One Yr data.csv',
    'INFY': 'Infosys one YR data.csv',
    'WIPRO': 'Wipro One Yr data.csv',
    'HCLTECH': 'HCLTech One Yr data.csv'
}

# Read the CSV files into DataFrames
data = {name: pd.read_csv(file) for name, file in files.items()}

# Process the data: Convert dates and set index
for name, df in data.items():
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    df.sort_index(inplace=True)

# Function to fit ARIMA model, forecast, and get confidence intervals
def forecast_arima(df, steps=30):
    model = ARIMA(df['Close'], order=(5, 1, 0))
    model_fit = model.fit()

    forecast_object = model_fit.get_forecast(steps=steps)
    forecast = forecast_object.predicted_mean
    conf_int = forecast_object.conf_int()

    return forecast, conf_int

# Prepare to plot all forecasts on the same graph
plt.figure(figsize=(14, 8))

# Forecast for each stock and plot on the same graph
for name, df in data.items():
    forecast, conf_int = forecast_arima(df)
    
    forecast_dates = pd.date_range(df.index[-1], periods=len(forecast), freq='D')
    plt.plot(forecast_dates, forecast, label=f'{name} Forecast')
    plt.fill_between(forecast_dates, conf_int.iloc[:, 0], conf_int.iloc[:, 1], alpha=0.1)

# Add titles and labels
plt.title('Stock Price Forecast Comparison with Confidence Intervals (Next 30 Days)')
plt.xlabel('Date')
plt.ylabel('Forecasted Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```
